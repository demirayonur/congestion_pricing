```python
import sys

sys.path.insert(1, '/Users/od721/Desktop/congestion_pricing/src')

import random_instance
import gurobi_base
import gurobi_base_fixed_variables

import pandas as pd
import numpy as np
```

```python
my_seed = 10
```

```python
options = [(2*i, 4*i, i) for i in range(10,41)]
```

```python
node_ls, arc_ls, com_ls, obj1_ls, obj2_ls = [], [], [], [], []
counter = 0
for opt in options:
    print(counter)
    counter += 1
    n_node, n_arc, n_commodity = opt
    our_instance = random_instance.Instance(n_node, n_arc, n_commodity, my_seed)
    obj_val, _, _ = gurobi_base.optimize_by_gurobi(our_instance,3600, log_output=False)
    obj_val2, _, _ = gurobi_base_fixed_variables.optimize_by_gurobi(our_instance,3600, log_output=False)
    node_ls.append(n_node)
    arc_ls.append(n_arc)
    com_ls.append(n_commodity)
    obj1_ls.append(round(obj_val,2))
    obj2_ls.append(round(obj_val2,2))
df = pd.DataFrame()
df['nNode'] = node_ls
df['nArc'] = arc_ls
df['nCom'] = com_ls
df['Obj1'] = obj1_ls
df['Obj2'] = obj2_ls
df['isEqual'] = df['Obj1'] == df['Obj2']
df
```
