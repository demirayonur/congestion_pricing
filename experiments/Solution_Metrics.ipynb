```python
import sys

sys.path.insert(1, '/Users/od721/Desktop/congestion_pricing/src')

import random_instance
from solution_metrics import optimize_with_solution_metrics as get_metrics

import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import math
import time
```

```python
seed = 10
n_node = 50
n_arc_ls = [int(math.ceil(n_node*(n_node-1)*(d/100))) for d in range(1,101,20)]
densities = [a/(n_node*(n_node-1))for a in n_arc_ls]
```

```python
densities
```

```python
com_ls, density_ls, rate_ls, constraint_util_ls = [], [], [], []
for c in range(1, 5):
    for i, n_arc in enumerate(n_arc_ls):
        print('c-' + str(c)+'_a-'+str(n_arc))
        com_ls.append(c)
        density_ls.append(densities[i])
        ins_ = random_instance.Instance(n_node, n_arc, c, seed)
        _, _, _, arc_util, num_eq, num_ineq, num_binding  = get_metrics(ins_, 3600, log_output=False)
        rate_ls.append(arc_util)
        constraint_util_ls.append(round((num_binding/num_ineq)*100,2))
        
```

```python
df = pd.DataFrame()
df['ncom'] = com_ls
df['density'] = density_ls
df['ColumnUtil'] = rate_ls
df['RowUtil'] = constraint_util_ls
```
